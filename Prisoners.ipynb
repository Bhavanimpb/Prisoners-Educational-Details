```python
import pandas as pd
df=pd.read_csv("prisoners.csv")
df
```

```python
df1=df.rename({"No. of Inmates benefitted by Elementary Education":"IEE","No. of Inmates benefitted by Adult Education":"IEA","No. of Inmates benefitted by Higher Education":"IHE","No. of Inmates benefitted by Computer Course":"ICC"},axis=1)
df1
```

```python
## add new column and total rows
df1=df1.assign(total_benefitted=df1.IEE+df1.IEA+df1.IHE+df1.ICC)
df1
```

```python
df1["total"]=df1.apply(lambda col:col["IEE"]+col["IEA"]+col["IHE"]+col["ICC"],axis=1)
df1
```

```python
x=df1["IEE"].sum(),df1["IEA"].sum(),df1["IHE"].sum(),df1["ICC"].sum()
x
```

```python
df=df.append(df.sum(),ignore_index=True)
df.at[35,"STATE/UT"]="Totals"
df.tail(5)
```

```python
## add new row and sum of columns
df2=df1.append(pd.Series(df1.sum(),name="total"))
df2
```
